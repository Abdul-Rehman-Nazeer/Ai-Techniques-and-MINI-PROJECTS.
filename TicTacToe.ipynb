```python
import random

class TicTacToe:
    def __init__(self):
        self.board = [[' ' for _ in range(3)] for _ in range(3)]
        self.current_player = 'X'  # Player is 'X', computer agent is 'O'

    def print_board(self):
        for row in self.board:
            print('|'.join(row))
            print('-' * 5)

    def is_winner(self, player):
        # Check rows for a win
        for i in range(3):
            if self.board[i][0] == self.board[i][1] == self.board[i][2] == player:
                return True

        # Check columns for a win
        for i in range(3):
            if self.board[0][i] == self.board[1][i] == self.board[2][i] == player:
                return True

        # Check diagonals for a win
        if self.board[0][0] == self.board[1][1] == self.board[2][2] == player:
            return True
        if self.board[0][2] == self.board[1][1] == self.board[2][0] == player:
            return True

        return False

    def is_draw(self):
        for row in self.board:
            for cell in row:
                if cell == ' ':
                    return False
        return True

    def player_move(self):
        while True:
            row = int(input("Enter row (0, 1, or 2): "))
            col = int(input("Enter column (0, 1, or 2): "))
            if self.board[row][col] == ' ':
                self.board[row][col] = self.current_player
                break
            else:
                print("Invalid move! Try again.")

    def computer_move(self):
        # Check for winning move for computer
        for row in range(3):
            for col in range(3):
                if self.board[row][col] == ' ':
                    self.board[row][col] = 'O'
                    if self.is_winner('O'):
                        return
                    else:
                        self.board[row][col] = ' '

        # Check for blocking move for player
        for row in range(3):
            for col in range(3):
                if self.board[row][col] == ' ':
                    self.board[row][col] = 'X'
                    if self.is_winner('X'):
                        self.board[row][col] = 'O'
                        return
                    else:
                        self.board[row][col] = ' '

        # If no winning or blocking move, make a random move
        while True:
            row = random.randint(0, 2)
            col = random.randint(0, 2)
            if self.board[row][col] == ' ':
                self.board[row][col] = 'O'
                return

    def play_game(self):
        print("Welcome to Tic Tac Toe!")
        print("You are playing as 'X'. Computer is playing as 'O'.")
        print("Enter row and column numbers (0, 1, or 2) to make your move.")
        self.print_board()

        while not self.is_winner('X') and not self.is_winner('O') and not self.is_draw():
            if self.current_player == 'X':
                self.player_move()
            else:
                self.computer_move()

            self.print_board()
            if self.current_player == 'X':
                self.current_player = 'O'
            else:
                self.current_player = 'X'

        if self.is_winner('X'):
            print("Congratulations! You win!")
        elif self.is_winner('O'):
            print("Computer wins! Better luck next time.")
        else:
            print("It's a draw! Nobody wins.")


game = TicTacToe()
game.play_game()
```
