NLTK

```python
# 1. Write a Python NLTK program to split the text sentence/paragraph into a list
# of words. Sample Below.

import nltk
from nltk.tokenize import sent_tokenize

data = "Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station."

sent = sent_tokenize(data)

print("Tokenized sentences as a list:")
print(sent)

print("\nSentences:")
for i in sent:
    print(i)
```

```python
# 2. Write a Python NLTK program to create a list of words from a given string.

import nltk
from nltk.tokenize import word_tokenize

data = "Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station."

list = word_tokenize(data)

print(list)
```

```python
# 3. Write a Python NLTK program to tokenize words, sentence wise.

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

data = "Joe waited for the train. The train was late. Mary and Samantha took the bus. I looked for Mary and Samantha at the bus station."

sent = sent_tokenize(data)

for i in sent:
    words = word_tokenize(i)
    print(words)

```

SPACY

```python
# 1. Create a syntactic dependency visualizer for a given sentence.

import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

sentence = "Joe waited for the train. The train was late."

doc = nlp(sentence)

displacy.render(doc, style="dep", jupyter=True)
```

```python
# 2. Break a given sentence into tokens each representing a single word.

import nltk
from nltk.tokenize import word_tokenize

sent = "I looked for Mary and Samantha at the bus station."

tokens = word_tokenize(sent)

print(tokens)

```
